# Data Cleaning

Data Type Fixes & Fixing Columns name

```python
# Fixing Column names
customers_df.columns
loans_df.columns
payments_df.columns
```

```python
# Fixing Data Types of customers_df
cat_cols = customers_df.select_dtypes('object').columns
cat_cols = cat_cols[1:]
for c in cat_cols :
  customers_df[c] = customers_df[c].astype('category')
customers_df.info()
```

```python
# Fixing Data Types of loans_df
loans_df['risk_category'] = loans_df['risk_category'].astype('category')
loans_df['loan_status'] = loans_df['loan_status'].astype('category')
loans_df['loan_purpose'] = loans_df['loan_purpose'].astype('category')

loans_df['sanctioned_date'] = pd.to_datetime(loans_df['sanctioned_date'])
loans_df['loan_end_date'] = pd.to_datetime(loans_df['loan_end_date'])
```

```python
# Fixing Data Types of payments_df
payments_df['payment_status'] = payments_df['payment_status'].astype('category')

payments_df['payment_date'] = pd.to_datetime(payments_df['payment_date'])
payments_df['due_date'] = pd.to_datetime(payments_df['due_date'])
```

```python
# Standardising text
cat_cols = customers_df.select_dtypes('category').columns
for c in cat_cols :
  customers_df[c] = customers_df[c].astype('str').str.strip().str.lower()
```

```python
cat_cols = loans_df.select_dtypes('category').columns
for c in cat_cols :
  loans_df[c] = loans_df[c].astype('str').str.strip().str.lower()
```

```python
cat_cols = payments_df.select_dtypes('category').columns
for c in cat_cols :
  payments_df[c] = payments_df[c].astype('str').str.strip().str.lower()
```

**Visual + Statistical Profiling For customers_df**

```python
# Identify categorical columns
cat_cols = customers_df.select_dtypes(include=['category', 'object']).columns

for col in cat_cols:
    print(f"\n==== {col.upper()} ====\n")

    # 1. Category Distribution (Chart)
    print("1. Category Distribution (Chart)\n")
    sns.countplot(x=customers_df[col])
    plt.xticks(rotation=45)
    plt.show()

    # 2. Category Distribution (Stats)
    print("\n2. Category Distribution (Stats)\n")
    print(customers_df[col].value_counts())
    print("\nPercentage Distribution:\n")
    print((customers_df[col].value_counts(normalize=True) * 100).round(2))

    # 3. Unique Categories
    print("\n3. Unique Categories\n")
    print("Unique Count:", customers_df[col].nunique())
    print("Categories:", list(customers_df[col].unique()))

    # 4. Missing Values (%)
    print("\n4. Missing Values (%)\n")
    nulls = customers_df[col].isna().sum()
    print((nulls / len(customers_df)) * 100)

    # 5. Rare Categories (less than 1%)
    print("\n5. Rare Categories (less than 1%)\n")
    rare = customers_df[col].value_counts(normalize=True)
    rare = rare[rare < 0.01]
    print(rare if len(rare) > 0 else "No rare categories")
```

```python
# Nulls
numeric_cols = customers_df.select_dtypes(include=['int64', 'float64']).columns

print("=== NULL VALUE CHECK ===\n")
for col in numeric_cols:
    nulls = customers_df[col].isna().sum()
    # Note: Replace len(customers_df) where the placeholder (*no of row*) was
    print(f"{col.upper()}: {nulls} Missing i.e ({(nulls/len(customers_df))*100:.2f}%)\n")
```

```python
# Outliers
print("=== OUTLIER CHECK ===\n")

for col in numeric_cols:
    # Plotting the boxplot to visualize outliers
    sns.boxplot(x=customers_df[col])
    plt.show()

    # Calculating IQR (Interquartile Range)
    Q1 = customers_df[col].quantile(0.25)
    Q3 = customers_df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Defining bounds for outliers
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Counting values outside the bounds
    outliers = ((customers_df[col] < lower) | (customers_df[col] > upper)).sum()

    print(f"\nOutlier : {col.upper()}: {outliers}\n")
    print('------------------------------------------------------------')
```

```python
print("=== SKEWNESS CHECK ===\n")

for col in numeric_cols:
    # Visualizing distribution with a histogram and KDE curve
    sns.histplot(customers_df[col], kde=True)
    plt.title(f"{col} Distribution")
    plt.show()

    # Calculating and rounding the skewness value
    skew_val = round(customers_df[col].skew(), 3)
    print(f"\n{col.upper()} Skewness: {skew_val}\n")
    print('------------------------------------------------------------')
```

# Data Validation

```python
# Check for age < 18 in customers_df
young_customers = customers_df[customers_df['age'] < 18]
if young_customers.empty:
    print('All customers have an age of 18 or greater.')
else:
    print(f'{len(young_customers)} customers have an age less than 18.')
    print('Here are some examples of violating rows:')
    display(young_customers.head())
```

```python
# Check for monthly_income < 0 in customers_df
negative_income_customers = customers_df[customers_df['monthly_income'] < 0]
if negative_income_customers.empty:
    print('All customers have a monthly income of 0 or greater.')
else:
    print(f'{len(negative_income_customers)} customers have a monthly income less than 0.')
    print('Here are some examples of violating rows:')
    display(negative_income_customers.head())
```

```python
# Check for loan_amount < 0 in loans_df
negative_loan_amount = loans_df[loans_df['loan_amount'] < 0]
if negative_loan_amount.empty:
    print('All loans have a loan amount of 0 or greater.')
else:
    print(f'{len(negative_loan_amount)} loans have a loan amount less than 0.')
    print('Here are some examples of violating rows:')
    display(negative_loan_amount.head())
```

```python
# Check for credit_score outside valid range (e.g., 300-850) in customers_df
invalid_credit_score_customers = customers_df[(customers_df['credit_score'] < 300) | (customers_df['credit_score'] > 850)]
if invalid_credit_score_customers.empty:
    print('All customers have a credit score within the valid range (300-850).')
else:
    print(f'{len(invalid_credit_score_customers)} customers have a credit score outside the valid range.')
    print('Here are some examples of violating rows:')
    display(invalid_credit_score_customers.head())
```

```python
# Check payment_date >= due_date in payments_df
incorrect_payment_dates = payments_df[payments_df['payment_date'] < payments_df['due_date']]
if incorrect_payment_dates.empty:
    print('All payment_date values are greater than or equal to due_date in payments_df.')
else:
    print(f'{len(incorrect_payment_dates)} rows violate the rule: payment_date < due_date.')
    print('Here are some examples of violating rows:')
    display(incorrect_payment_dates.head())
```

```python
# Check remaining_balance >= 0 in payments_df
negative_remaining_balance = payments_df[payments_df['remaining_balance'] < 0]
if negative_remaining_balance.empty:
    print('All remaining_balance values are greater than or equal to 0 in payments_df.')
else:
    print(f'{len(negative_remaining_balance)} rows violate the rule: remaining_balance < 0.')
    print('Here are some examples of violating rows:')
    display(negative_remaining_balance.head())
```

```python
# Check loan_end_date >= sanctioned_date in loans_df
incorrect_loan_dates = loans_df[loans_df['loan_end_date'] < loans_df['sanctioned_date']]
if incorrect_loan_dates.empty:
    print('All loan_end_date values are greater than or equal to sanctioned_date in loans_df.')
else:
    print(f'{len(incorrect_loan_dates)} rows violate the rule: loan_end_date < sanctioned_date.')
    print('Here are some examples of violating rows:')
    display(incorrect_loan_dates.head())
```

```python
# Check installment_number <= total_installments
# Merge payments_df with loans_df to access total_installments
merged_df = pd.merge(payments_df, loans_df[['loan_id', 'total_installments']], on='loan_id', how='left')

incorrect_installments = merged_df[merged_df['installment_number'] > merged_df['total_installments']]

if incorrect_installments.empty:
    print('All installment_number values are less than or equal to total_installments.')
else:
    print(f'{len(incorrect_installments)} rows violate the rule: installment_number > total_installments.')
    print('Here are some examples of violating rows:')
    display(incorrect_installments[['loan_id', 'installment_number', 'total_installments']].head())
```

```python
print('Checking if all customer_ids in loans_df are present in customers_df:')
loan_customer_ids_not_in_customers = loans_df[~loans_df['customer_id'].isin(customers_df['customer_id'])]
if loan_customer_ids_not_in_customers.empty:
    print('All customer_ids in loans_df exist in customers_df.')
else:
    print(f'{len(loan_customer_ids_not_in_customers)} customer_ids in loans_df do not exist in customers_df.')
    display(loan_customer_ids_not_in_customers['customer_id'].unique())
```

```python
print('\nChecking if all loan_ids in payments_df are present in loans_df:')
payment_loan_ids_not_in_loans = payments_df[~payments_df['loan_id'].isin(loans_df['loan_id'])]
if payment_loan_ids_not_in_loans.empty:
    print('All loan_ids in payments_df exist in loans_df.')
else:
    print(f'{len(payment_loan_ids_not_in_loans)} loan_ids in payments_df do not exist in loans_df.')
    display(payment_loan_ids_not_in_loans['loan_id'].unique())
```

# Feature Engineering

```python
loans_df['loan_duration_days'] = (loans_df['loan_end_date']-loans_df['sanctioned_date']).dt.days
```

```python
payments_df['is_late_payment'] =  (payments_df['payment_delay_days']>0).astype(int)
```

```python
payments_df['is_missed_payment'] = (payments_df['payment_status'] == "missed").astype(int)
```

```python
loan_delay_summary = (payments_df.groupby('loan_id')['payment_delay_days'].sum().reset_index())
```

```python
loans_df['is_default'] = (loans_df['loan_status']=="defaulted").astype(int)
```

```python
payments_df.groupby('is_late_payment')['payment_id'].count()
```

```python
payments_df.groupby('is_missed_payment')['payment_id'].count()
```

```python
loans_df.groupby('is_default')['loan_id'].count()
```

```python
missed_per_loan = (
    payments_df
    .groupby('loan_id')['is_missed_payment']
    .sum()
    .reset_index(name='total_missed_payments')
)
missed_per_loan
```

```python
late_ratio = (
    payments_df
    .groupby('loan_id')['is_late_payment']
    .mean()
    .reset_index(name='late_payment_ratio')
)
late_ratio
```

```python
loans = loans_df.merge(customers_df, on='customer_id')

loans['debt_to_income_ratio'] = (
    loans['monthly_installment'] /
    customers['monthly_income']
)
loans
```

```python
loans = loans.merge(missed_per_loan, on='loan_id', how='left')
loans_defaulted = loans[loans['is_default'] == 1]

total_defaulted_loans = loans_defaulted.shape[0]
avg_missed_payments_per_defaulted_loan = loans_defaulted['total_missed_payments'].mean()
avg_debt_to_income_ratio_defaulted_loans = loans_defaulted['debt_to_income_ratio'].mean()

print(f"Total defaulted loans: {total_defaulted_loans}")
print(f"Average missed payments per defaulted loan: {avg_missed_payments_per_defaulted_loan:.2f}")
print(f"Average debt-to-income ratio of defaulted loans: {avg_debt_to_income_ratio_defaulted_loans:.2f}")
```
